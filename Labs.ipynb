Створимо схему для датафрейму Студенти. Створимо дані для датафреймів.
Також створимо самі датафрейми

```python
Myschema = StructType([
    StructField("Ім'я студента", StringType(), False),
    StructField("Прізвище студента", StringType(), False),
    StructField("Група студента", StringType(), False),
    StructField("Кафедра студента", StringType(), False)
])

data_it = [
    ("Анастасія", "Петренко", "ІТ-21", "Кафедра інформаційних технологій"),
    ("Олексій", "Іванов", "ІТ-21", "Кафедра інформаційних технологій"),
    ("Марія", "Шевченко", "ІТ-22", "Кафедра інформаційних технологій"),
    ("Ігор", "Коваленко", "ІТ-23", "Кафедра інформаційних технологій")
]

data_economics = [
    ("Олена", "Ткаченко", "ФЕ-21", "Кафедра економіки"),
    ("Дмитро", "Климов", "ФЕ-22", "Кафедра економіки"),
    ("Вікторія", "Мельник", "ФЕ-23", "Кафедра економіки"),
    ("Сергій", "Поліщук", "ФЕ-24", "Кафедра економіки")
]
```

```python
df_it = spark.createDataFrame(data_it, Myschema)
df_economics = spark.createDataFrame(data_economics, Myschema)
df_it.printSchema()
df_economics.printSchema()
```

Об'єднаємо два датафрейми в один

```python
df_university = df_it.union(df_economics)
df_university.show()
```

Додамо нову колонку

```python
df_university=df_university.withColumn("Бюджетна форма навчання", when(rand() < 0.5, True).otherwise(False))
df_university.show()
```

Додамо новий рядок

```python
df_university=df_university.union(spark.createDataFrame([("Іван", "Григоренко", "ІТ-24", "Кафедра інформаційних технологій", False)], df_university.schema))
df_university.show()
```

Видалимо рядок

```python
df_university = df_university.filter(df_university["Група студента"]!="ІТ-24")
df_university.show()
```

Видалимо стовпець

```python
df_university = df_university.drop("Кафедра студента")
df_university.show()
```

```python
spark.stop()
```

```python
spark = SparkSession.builder.appName("Anastasiia Lytvyn IS-42mp LAB-1 Possum").getOrCreate()
```

Створити свою схему та завантажити дані з файлу, виконати дії
відповідно до варіанту.

```python
Myschema = StructType([
    StructField("", IntegerType(), True),
    StructField("case", IntegerType(), True),
    StructField("site", IntegerType(), True),
    StructField("Pop", StringType(), True),
    StructField("sex", StringType(), True),
    StructField("age", DoubleType(), True),
    StructField("hdlngth", DoubleType(), True),
    StructField("skullw", DoubleType(), True),
    StructField("totlngth", DoubleType(), True),
    StructField("taill", DoubleType(), True),
    StructField("footlgth", DoubleType(), True),
    StructField("earconch", DoubleType(), True),
    StructField("eye", DoubleType(), True),
    StructField("chest", DoubleType(), True),
    StructField("belly", DoubleType(), True)
])
```

```python
df=spark.read.csv("/content/possum.csv", header=True, schema=Myschema)
df = df.drop("")
df.printSchema()
```

```python
df.show(105)
```

1. Вивести вік всіх самиць з провінції Вікторія.

```python
df.select("age").filter( (col("Pop") == "Vic") & (col("sex") == "f") ).show(105)
```

2. Порахувати загальну популяцію опосумів усіх провінцій, крім
Вікторії.

```python
df.filter(col("Pop") != "Vic").count()
```

3. Порахувати середню довжину хвоста самців.

```python
df.filter(col("sex") == "m").agg({"taill": "avg"}).show()
```

4. Відсортувати дані за загальною довжиною тіла (спадаючий
порядок); за віком (зростаючий).

```python
df.orderBy(col("totlngth").desc(), col("age").asc()).show(50)
```

```python
df.orderBy(col("age").asc(),col("totlngth").desc()).show(50)
```
